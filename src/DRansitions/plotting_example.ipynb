```python
from xSM import *
import matplotlib.pyplot as plt
from matplotlib import rc,rcParams

rcParams['font.size'] = 18
rcParams['font.family'] = 'serif'
```

```python
#specify order of potential and EFT
order_potential = 1
order_EFT = 2

#specify 3D RG scale in units of T, 4D RG scale in units of 4 * pi * exp(-euler_gamma) 
mu3Bar = 1
mu4D = 1

#singlet parameters
mh2 = 100
lambdaMix = 0.86
lambdaS = 1

#One-loop corrected 4D input parameters at scale mu = M_Z for above singlet parameters (see appendix of 2103.07467)
g = 0.6484344753719615
gPrime = 0.35673693719015237
yt = 0.9739993842413421
mPhiSq = -8557.903480000112
mSSq = -16698.670212217236
lambdaPhi = 0.14368503519337655
gSSq = 1.48409

inputparams = [g,gPrime,yt,mPhiSq,mSSq,lambdaPhi,gSSq]
```

```python
xSM_Class = xSM(mh2,lambdaMix,lambdaS,inputparams,mu3Bar,mu4D,order_EFT,order_potential)
```

```python
listLen=1000

#Input: 4D Fields
phi = np.linspace(1e-10,250,listLen)
s = 0

Fields = np.array([[phi[i],s] for i in range(len(phi))])

#set temperature in GeV
T0 = 200
V0 = xSM_Class.Vtot(Fields,T0)

T1 = 100
V1 = xSM_Class.Vtot(Fields,T1)
```

```python
#Potential in phi-direction

fig, ax = plt.subplots(figsize=(9,6))
ax.plot(phi, V0-V0[0], lw=3, label=r'$T = 200$ GeV')
ax.plot(phi, V1-V1[0], lw=3, label=r'$T = 100$ GeV')

ax.set_xlabel(r'$h_\mathrm{4D}$ [GeV]',labelpad=10, fontsize=22)
ax.set_ylabel(r'$T\;V_\mathrm{eff,3D}$ [GeV]', fontsize=22)
ax.set_xlim(0,250)
ax.grid(alpha=.3)
ax.legend()
```
